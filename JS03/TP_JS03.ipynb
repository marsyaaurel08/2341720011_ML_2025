```python
# Tugas JS03
# 1. Membaca dataset
file_path = '/content/drive/MyDrive/Machine Learning   15/Data/wbc.csv'
df = pd.read_csv(file_path)
```

```python
# 2. Menghapus kolom yang tidak digunakan (id dan kolom kosong Unnamed: 32)
df_clean = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
# 3. Encoding kolom diagnosis (M -> 1, B -> 0)
df_clean['diagnosis'] = df_clean['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# 4. Standarisasi semua kolom numerik kecuali diagnosis
scaler = StandardScaler()
numerical_cols = df_clean.columns.drop('diagnosis')
df_clean[numerical_cols] = scaler.fit_transform(df_clean[numerical_cols])
```

```python
# 5. Memisahkan fitur (X) dan target (y)
X = df_clean.drop(columns=['diagnosis'])
y = df_clean['diagnosis']
```

```python
# 6. Stratified Split (80:20) agar distribusi kelas seimbang
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# 7. Menampilkan informasi hasil
print("Jumlah data awal:", len(df))
print("Jumlah data setelah cleaning:", len(df_clean))
print("Jumlah kolom yang digunakan:", len(df_clean.columns))
print("Jumlah data latih:", len(X_train))
print("Jumlah data uji:", len(X_test))
print("Distribusi kelas data latih:\n", y_train.value_counts())
print("Distribusi kelas data uji:\n", y_test.value_counts())
```
