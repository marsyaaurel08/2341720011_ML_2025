```python
# Age - mean
df_from_drive['Age'].fillna(value=df_from_drive['Age'].mean(), inplace=True)

# Cabin - "DECK"
df_from_drive['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df_from_drive['Embarked'].fillna(value=df_from_drive['Embarked'].mode, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# === EKSTRAKSI FITUR ===

# 1. FamilySize: jumlah keluarga di kapal (SibSp + Parch + diri sendiri)
df_from_drive["FamilySize"] = df_from_drive["SibSp"] + df_from_drive["Parch"] + 1

# 2. Title: ambil gelar dari kolom Name (Mr, Mrs, Miss, Master, dll.)
df_from_drive["Title"] = df_from_drive["Name"].str.extract(r',\s*([^\.]+)\.', expand=False)
title_mapping = {
    "Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs",
    "Lady": "Rare", "Countess": "Rare", "Capt": "Rare", "Col": "Rare",
    "Don": "Rare", "Dr": "Rare", "Major": "Rare", "Rev": "Rare",
    "Sir": "Rare", "Jonkheer": "Rare", "Dona": "Rare"
}
df_from_drive["Title"] = df_from_drive["Title"].replace(title_mapping)

# 3. AgeBin: kelompokkan usia menjadi kategori
df_from_drive["AgeBin"] = pd.cut(df_from_drive["Age"], bins=[0, 12, 18, 35, 60, 100],
                      labels=["Child", "Teen", "YoungAdult", "Adult", "Senior"])

# 4. CabinDeck: ambil huruf pertama dari kolom Cabin
df_from_drive["CabinDeck"] = df_from_drive["Cabin"].astype(str).str[0]

# 5. Fare per person: bagi harga tiket dengan jumlah keluarga
df_from_drive["FarePerPerson"] = df_from_drive["Fare"] / df_from_drive["FamilySize"]

print(" Hasil Ekstraksi fitur ")
df_from_drive[["Name", "Title", "Age", "AgeBin", "FamilySize", "Cabin", "CabinDeck", "FarePerPerson"]].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# === VISUALISASI FITUR HASIL EKSTRAKSI ===

plt.figure(figsize=(12, 6))
sns.barplot(x="FamilySize", y="Survived", data=df_from_drive)
plt.title("Survival Rate berdasarkan Family Size")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="Title", y="Survived", data=df_from_drive,
            order=df_from_drive.groupby("Title")["Survived"].mean().sort_values().index)
plt.title("Survival Rate berdasarkan Title")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="AgeBin", y="Survived", data=df_from_drive,
            order=["Child", "Teen", "YoungAdult", "Adult", "Senior"])
plt.title("Survival Rate berdasarkan Kelompok Usia (AgeBin)")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="CabinDeck", y="Survived", data=df_from_drive,
            order=df_from_drive.groupby("CabinDeck")["Survived"].mean().sort_values().index)
plt.title("Survival Rate berdasarkan Cabin Deck")
plt.show()
```
